```python
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.keys import Keys
```

```python
olx=webdriver.Chrome("chromedriver.exe")
olx.get("https://www.olx.in/")
select=olx.find_element_by_xpath("//input[@data-aut-id='searchBox']")
select.send_keys(Keys.CONTROL+ "a",Keys.DELETE)
select.send_keys("Cars",Keys.ENTER)
time.sleep(3)
olx.find_element_by_xpath("//*[@class='_1-yWE rui-2THJR rui-A9TyJ ']").click()
time.sleep(4)
olx.find_element_by_xpath("//*[@class='_2VpYb'][contains(text(),'Date Published')]").click()
time.sleep(5)
for i in range(1000):
    try:
        select=olx.find_element_by_xpath("html ")
        select.send_keys(Keys.END)
        time.sleep(4)
        olx.find_element_by_xpath("//*[contains(text(),'load more')]").click()
        time.sleep(4)
    except:
        break
#all cars:
link=olx.find_elements_by_xpath("//li[@class='EIR5N']//a")
url=[]
for l in link:
    url.append(l.get_attribute("href"))
brand=[]
model=[]
varient=[]
fuel=[]
year=[]
transmission=[]
km=[]
nod=[]
location=[]
price=[]
for u in url:
    olx.get(u)
    #brand
    try:
        brand.append(olx.find_element_by_xpath("//*[contains(text(),'Brand')]/../*[@class='_2vNpt']").text)
    except:
        brand.append("nan")
    #model
    try:
        model.append(olx.find_element_by_xpath("//*[contains(text(),'Model')]/../*[@class='_2vNpt']").text)
    except:
        model.append("nan")
    #varient
    try:
        varient.append(olx.find_element_by_xpath("//*[contains(text(),'Variant')]/../*[@class='_2vNpt']").text)
    except:
        varient.append("nan")
    #fuel
    try:
        fuel.append(olx.find_element_by_xpath("//*[contains(text(),'Fuel')]/../*[@class='_2vNpt']").text)
    except:
        fuel.append("nan")
    #year
    try:
        year.append(olx.find_element_by_xpath("//*[contains(text(),'Year')]/../*[@class='_2vNpt']").text)
    except:
        year.append("nan")
    #transmission
    try:
        transmission.append(olx.find_element_by_xpath("//*[contains(text(),'Transmission')]/../*[@class='_2vNpt']").text)
    except:
        transmission.append("nan")
    #km
    try:
        km.append(olx.find_element_by_xpath("//*[contains(text(),'KM driven')]/../*[@class='_2vNpt']").text)
    except:
        km.append("nan")
    try:
        nod.append(olx.find_element_by_xpath("//*[contains(text(),'No. of Owners')]/../*[@class='_2vNpt']").text)
    except:
        nod.append("nan")
    #location
    try:
        location.append(olx.find_element_by_xpath("//*[@class='_2FRXm']").text)
    except:
        location.append("nan")
    #price
    try:
        price.append(olx.find_element_by_xpath("//*[@class='_2xKfz']").text)
    except:
        price.append("nan")
```

```python
data={"brand":brand,"nod":nod,"location":location,"model":model,"varient":varient,"fuel":fuel,"year":year,"transmission":transmission,"km":km,"price":price}
df2=pd.DataFrame(data)
```

```python
for i in df2.columns:
    ind=df2[df2[i]=="nan"].index
    df2.drop(ind,inplace=True)
```

```python
df2.to_csv("olx")
```
